```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Define the parametric line: R(t) = R_0 + t*v
# R_0 is the initial position vector (point on the line)
# v is the direction vector
R_0 = np.array([1, 2, 3])  # Starting point
v = np.array([2, 1, 1.5])  # Direction vector

# Parameter t values
t_values = np.linspace(-1, 3, 100)

# Calculate points along the line
line_points = np.array([R_0 + t * v for t in t_values])

# Specific points for annotation
t_0 = 0  # Point P_0
t_p = 1.5  # Point P
P_0 = R_0 + t_0 * v
P = R_0 + t_p * v
R = P - np.array([0, 0, 0])  # Position vector to P

# Create the figure
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot the line L
ax.plot(line_points[:, 0], line_points[:, 1], line_points[:, 2], 
        color='darkred', linewidth=3, label='Line L')

# Origin
origin = np.array([0, 0, 0])

# Plot R_0 (position vector from origin to P_0)
ax.quiver(origin[0], origin[1], origin[2], 
          R_0[0], R_0[1], R_0[2],
          color='darkred', arrow_length_ratio=0.15, linewidth=2.5)
ax.text(R_0[0]/2 - 0.3, R_0[1]/2 - 0.3, R_0[2]/2, r'$\mathbf{R_0}$', 
        fontsize=14, color='darkred', fontweight='bold')

# Plot R (position vector from origin to P)
ax.quiver(origin[0], origin[1], origin[2], 
          P[0], P[1], P[2],
          color='darkred', arrow_length_ratio=0.1, linewidth=2.5)
ax.text(P[0]/2 + 0.2, P[1]/2, P[2]/2 - 0.3, r'$\mathbf{R}$', 
        fontsize=14, color='darkred', fontweight='bold')

# Plot direction vector v from P
ax.quiver(P[0], P[1], P[2], 
          v[0], v[1], v[2],
          color='darkred', arrow_length_ratio=0.2, linewidth=2.5)
ax.text(P[0] + v[0]/2 + 0.3, P[1] + v[1]/2 + 0.2, P[2] + v[2]/2, 
        r'$\mathbf{v}$', fontsize=14, color='darkred', fontweight='bold')

# Mark points P_0 and P
ax.scatter(*P_0, color='darkred', s=100, zorder=5)
ax.text(P_0[0] - 0.2, P_0[1] - 0.3, P_0[2] + 0.3, r'$P_0$', 
        fontsize=13, color='black', fontweight='bold')

ax.scatter(*P, color='darkred', s=100, zorder=5)
ax.text(P[0] - 0.2, P[1] + 0.3, P[2] + 0.3, r'$P$', 
        fontsize=13, color='black', fontweight='bold')

# Mark origin
ax.scatter(*origin, color='black', s=100, zorder=5)
ax.text(origin[0] - 0.5, origin[1] - 0.3, origin[2] - 0.3, r'$O$', 
        fontsize=13, color='black', fontweight='bold')

# Label the line
line_label_point = line_points[10]
ax.text(line_label_point[0] - 0.5, line_label_point[1], 
        line_label_point[2] + 0.3, r'$L$', 
        fontsize=14, color='darkred', fontweight='bold')

# Draw axes through origin
axis_length = 6
ax.plot([0, axis_length], [0, 0], [0, 0], 'k-', linewidth=1, alpha=0.3)
ax.plot([0, 0], [0, axis_length], [0, 0], 'k-', linewidth=1, alpha=0.3)
ax.plot([0, 0], [0, 0], [0, axis_length], 'k-', linewidth=1, alpha=0.3)

# Set labels
ax.set_xlabel('x', fontsize=12)
ax.set_ylabel('y', fontsize=12)
ax.set_zlabel('z', fontsize=12)

# Set viewing angle
ax.view_init(elev=20, azim=130)

# Set equal aspect ratio and limits
max_range = 6
ax.set_xlim([0, max_range])
ax.set_ylim([0, max_range])
ax.set_zlim([0, max_range])

# Add title
ax.set_title('Parametric/Vector Equation of a Line in 3D Space\n' + 
             r'$\mathbf{R}(t) = \mathbf{R_0} + t\mathbf{v}$', 
             fontsize=14, fontweight='bold', pad=20)

# Add grid
ax.grid(True, alpha=0.3)

plt.tight_layout()
plt.show()

# Print the parametric equations
print("Parametric Equation of Line L:")
print(f"R(t) = R_0 + t*v")
print(f"R(t) = {R_0} + t*{v}")
print(f"\nIn component form:")
print(f"x(t) = {R_0[0]} + {v[0]}t")
print(f"y(t) = {R_0[1]} + {v[1]}t")
print(f"z(t) = {R_0[2]} + {v[2]}t")
print(f"\nPoint P_0 (at t=0): {P_0}")
print(f"Point P (at t={t_p}): {P}")
print(f"Direction vector v: {v}")
```

Skew Lines:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# First line: Passes through point A and its direction vector is D1
# R1 = A + t*D1
A = np.array([1, 0, 0])
D1 = np.array([-1, 1, 0])

# Second line: Passes through point B and its direction vector is D2
# R2 = B + s*D2
B = np.array([0, 1, 1])
D2 = np.array([0, 1, 1])

# Generate parameter values for lines
t = np.linspace(-5, 5, 100)  # Parameter for lines

# Generate points for each line
line1_x = A[0] + t * D1[0]
line1_y = A[1] + t * D1[1]
line1_z = A[2] + t * D1[2]

line2_x = B[0] + t * D2[0]
line2_y = B[1] + t * D2[1]
line2_z = B[2] + t * D2[2]

# Create the figure
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot the lines
ax.plot(line1_x, line1_y, line1_z, color='green', linewidth=2.5, label='Line 1')
ax.plot(line2_x, line2_y, line2_z, color='red', linewidth=2.5, label='Line 2')

# Mark points A and B
ax.scatter(*A, color='green', s=100, zorder=5)
ax.text(A[0] + 0.2, A[1] - 0.3, A[2], r'$\mathbf{A}$', 
        fontsize=13, color='black', fontweight='bold')

ax.scatter(*B, color='red', s=100, zorder=5)
ax.text(B[0] - 0.3, B[1], B[2] + 0.2, r'$\mathbf{B}$', 
        fontsize=13, color='black', fontweight='bold')

# Direction vectors
# D1 from point A
ax.quiver(A[0], A[1], A[2], 
          D1[0], D1[1], D1[2],
          color='green', arrow_length_ratio=0.2, linewidth=2.5)
ax.text(A[0] + D1[0]/2 - 0.4, A[1] + D1[1]/2 + 0.2, A[2] + D1[2]/2, 
        r'$\mathbf{D_1}$', fontsize=12, color='green', fontweight='bold')

# D2 from point B
ax.quiver(B[0], B[1], B[2], 
          D2[0], D2[1], D2[2],
          color='red', arrow_length_ratio=0.2, linewidth=2.5)
ax.text(B[0] + D2[0]/2 + 0.1, B[1] + D2[1]/2 + 0.2, B[2] + D2[2]/2, 
        r'$\mathbf{D_2}$', fontsize=12, color='red', fontweight='bold')

# Set labels
ax.set_xlabel('x-axis', fontsize=11)
ax.set_ylabel('y-axis', fontsize=11)
ax.set_zlabel('z-axis', fontsize=11)

# Set viewing angle similar to the image
ax.view_init(elev=15, azim=45)

# Set limits
ax.set_xlim([-4, 4])
ax.set_ylim([-2, 6])
ax.set_zlim([-2, 6])

# Add title
ax.set_title('Skew Lines in 3D Space', fontsize=14, fontweight='bold', pad=20)

# Add legend
ax.legend(loc='upper right', fontsize=10)

# Add grid
ax.grid(True, alpha=0.3)

# Add text box with line equations
param_text = (
    f'$\\mathbf{{A}} = ({A[0]}, {A[1]}, {A[2]})$\n'
    f'$\\mathbf{{D_1}} = ({D1[0]}, {D1[1]}, {D1[2]})$\n'
    f'$\\mathbf{{B}} = ({B[0]}, {B[1]}, {B[2]})$\n'
    f'$\\mathbf{{D_2}} = ({D2[0]}, {D2[1]}, {D2[2]})$'
)

# Position text box in figure coordinates
fig.text(0.15, 0.75, param_text, fontsize=11, 
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8, edgecolor='black'))

plt.tight_layout()
plt.show()

# Print the parametric equations
print("Skew Lines in 3D Space")
print("=" * 50)
print("\nFirst Line: Passes through point A and its direction vector is D1")
print(f"A = {A}")
print(f"D1 = {D1}")
print(f"R1(t) = A + t*D1")
print(f"R1(t) = {A} + t*{D1}")
print(f"\nParametric form:")
print(f"  x(t) = {A[0]} + {D1[0]}*t")
print(f"  y(t) = {A[1]} + {D1[1]}*t")
print(f"  z(t) = {A[2]} + {D1[2]}*t")

print("\n" + "=" * 50)
print("\nSecond Line: Passes through point B and its direction vector is D2")
print(f"B = {B}")
print(f"D2 = {D2}")
print(f"R2(s) = B + s*D2")
print(f"R2(s) = {B} + s*{D2}")
print(f"\nParametric form:")
print(f"  x(s) = {B[0]} + {D2[0]}*s")
print(f"  y(s) = {B[1]} + {D2[1]}*s")
print(f"  z(s) = {B[2]} + {D2[2]}*s")

# Check if lines are skew (not parallel and not intersecting)
print("\n" + "=" * 50)
print("\nSkew Lines Properties:")
print("- Not parallel (direction vectors not scalar multiples)")
print("- Do not intersect (no common point)")
print("- Exist in different planes in 3D space")
```

Parallel Lines:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# First line: Passes through point A and its direction vector is D1
# R1 = A + t*D1
A = np.array([1, 0, 0])
D1 = np.array([1, 1, 0])

# Second line: Passes through point B and its direction vector is D2
# R2 = B + s*D2
# For parallel lines, D2 must be a scalar multiple of D1
B = np.array([0, 1, 1])
D2 = np.array([2, 2, 0])  # D2 = 2*D1, so lines are parallel

# Generate parameter values for lines
t = np.linspace(-5, 5, 100)  # Parameter for lines

# Generate points along the lines for visualization
line1_x = A[0] + t * D1[0]
line1_y = A[1] + t * D1[1]
line1_z = A[2] + t * D1[2]

line2_x = B[0] + t * D2[0]
line2_y = B[1] + t * D2[1]
line2_z = B[2] + t * D2[2]

# Create the 3D plot
fig = plt.figure(figsize=(12, 9))
ax = fig.add_subplot(111, projection='3d')

# Plot the lines
ax.plot(line1_x, line1_y, line1_z, color='green', linewidth=2.5, label='Line 1')
ax.plot(line2_x, line2_y, line2_z, color='red', linewidth=2.5, label='Line 2')

# Mark points A and B
ax.scatter(*A, color='green', s=100, zorder=5)
ax.text(A[0] + 0.3, A[1] - 0.3, A[2], r'$\mathbf{A}$', 
        fontsize=13, color='black', fontweight='bold')

ax.scatter(*B, color='red', s=100, zorder=5)
ax.text(B[0] - 0.3, B[1] + 0.2, B[2] + 0.3, r'$\mathbf{B}$', 
        fontsize=13, color='black', fontweight='bold')

# Direction vectors
# D1 from point A
ax.quiver(A[0], A[1], A[2], 
          D1[0], D1[1], D1[2],
          color='green', arrow_length_ratio=0.2, linewidth=2.5)
ax.text(A[0] + D1[0]/2 + 0.2, A[1] + D1[1]/2 + 0.2, A[2] + D1[2]/2, 
        r'$\mathbf{D_1}$', fontsize=12, color='green', fontweight='bold')

# D2 from point B
ax.quiver(B[0], B[1], B[2], 
          D2[0]*0.5, D2[1]*0.5, D2[2]*0.5,  # Scale down for visualization
          color='red', arrow_length_ratio=0.2, linewidth=2.5)
ax.text(B[0] + D2[0]/4 + 0.3, B[1] + D2[1]/4 + 0.2, B[2] + D2[2]/4, 
        r'$\mathbf{D_2}$', fontsize=12, color='red', fontweight='bold')

# Set labels
ax.set_xlabel('x-axis', fontsize=11)
ax.set_ylabel('y-axis', fontsize=11)
ax.set_zlabel('z-axis', fontsize=11)

# Set viewing angle
ax.view_init(elev=20, azim=45)

# Set limits
ax.set_xlim([-5, 10])
ax.set_ylim([-5, 10])
ax.set_zlim([0, 1.5])

# Add title
ax.set_title('Parallel Lines in 3D Space', fontsize=14, fontweight='bold', pad=20)

# Add legend
ax.legend(loc='upper right', fontsize=10)

# Add grid
ax.grid(True, alpha=0.3)

# Add text box with line equations
param_text = (
    f'$\\mathbf{{A}} = ({A[0]}, {A[1]}, {A[2]})$\n'
    f'$\\mathbf{{D_1}} = ({D1[0]}, {D1[1]}, {D1[2]})$\n'
    f'$\\mathbf{{B}} = ({B[0]}, {B[1]}, {B[2]})$\n'
    f'$\\mathbf{{D_2}} = ({D2[0]}, {D2[1]}, {D2[2]})$'
)

# Position text box in figure coordinates
fig.text(0.15, 0.75, param_text, fontsize=11, 
         bbox=dict(boxstyle='round', facecolor='white', alpha=0.8, edgecolor='black'))

plt.tight_layout()
plt.show()

# Print the parametric equations
print("Parallel Lines in 3D Space")
print("=" * 50)
print("\nFirst line: Passes through point A and its direction vector is D1")
print(f"A = {A}")
print(f"D1 = {D1}")
print(f"R1(t) = A + t*D1")
print(f"R1(t) = {A} + t*{D1}")
print(f"\nParametric form:")
print(f"  x(t) = {A[0]} + {D1[0]}*t")
print(f"  y(t) = {A[1]} + {D1[1]}*t")
print(f"  z(t) = {A[2]} + {D1[2]}*t")

print("\n" + "=" * 50)
print("\nSecond line: Passes through point B and its direction vector is D2")
print(f"B = {B}")
print(f"D2 = {D2}")
print(f"R2(s) = B + s*D2")
print(f"R2(s) = {B} + s*{D2}")
print(f"\nParametric form:")
print(f"  x(s) = {B[0]} + {D2[0]}*s")
print(f"  y(s) = {B[1]} + {D2[1]}*s")
print(f"  z(s) = {B[2]} + {D2[2]}*s")

# Check if lines are parallel
print("\n" + "=" * 50)
print("\nParallel Lines Properties:")
print(f"D2 = {D2[0]/D1[0]:.1f} * D1")
print("Direction vectors are scalar multiples of each other")
print("Lines never intersect and maintain constant distance")
print("Lines lie in the same plane or parallel planes")
```
